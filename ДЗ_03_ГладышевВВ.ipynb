```python
rcParams["figure.figsize"] = 3, 3
plt.axis([-3, 3, -3, 3])
pylab.xticks(range(-3,3))
plt.grid(color="lightgrey")

x = []
y1 = []
y2 = []
r = 2
for i in range (-200, 201):
    x_tmp = i/100
    x.append(x_tmp)
    y1.append(r*math.sqrt(r**2-x_tmp**2)/r)
    y2.append(-r*math.sqrt(r**2-x_tmp**2)/r)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def pol_dec(a, r):
    x = r*cos(a)
    y = r*sin(a)
    return x, y
```

```python
polar_a = [np.pi, np.pi/2, np.pi/4]
polar_r = [1, 4, 3]

for ind in range(len(polar_a)):
    x, y = pol_dec(polar_a[ind], polar_r[ind])
    print("x: {:.3f},\ty: {:.3f}".format(x, y))
```

```python
rcParams["figure.figsize"] = 5, 5
plt.grid(color="lightgrey")

x = np.linspace(-5, 5, 21)

y1 = x**2-1
y2 = (exp(x) + x - 1)/x

#plt.axis([-5, 5, -5, 5])
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
from scipy.optimize import fsolve

def eq_1(p):
    x, y = p
    return (x**2-y-1, exp(x)+x*(1-y)-1)

x_tst = [-2, 1, 2, 5]
y_tst = [1, 1, 1, 1]

for ind in range(len(x_tst)):
    x_t, y_t = fsolve(eq_1, (x_tst[ind], y_tst[ind]))
    print("{} x: {:.3f},\ty: {:.3f}".format(ind+1,x_t, y_t))
    
```

### система 2

```python
rcParams["figure.figsize"] = 5, 5
plt.grid(color="lightgrey")

x = np.linspace(-6, 6, 50)
sol = []

for i in x:
    t = i**2-1
    if t > (exp(i) + i - 1)/i:
        sol.append(t)
    else:
        sol.append(np.NaN)
    
plt.plot(sol)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Решения системы с неравенством")
plt.show()
```
